```python
def eh_bissexto(ano):
  return (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)

def dias_no_mes(mes, ano):
    if mes in [4, 6, 9, 11]:
        return 30
    elif mes == 2:
        return 29 if eh_bissexto(ano) else 28
    else:
        return 31

def dias_desde_inicio_ano(dia, mes, ano):
    dias = 0
    for m in range(1, mes):
        dias += dias_no_mes(m, ano)
    dias += dia
    return dias

def calcular_diferenca(dia_aniv, mes_aniv, dia_atual, mes_atual):
    ano_atual = 2025  # Ano fixo, pode ser ajustado para ser dinâmico
    dias_ano = 366 if eh_bissexto(ano_atual) else 365

    dias_aniversario = dias_desde_inicio_ano(dia_aniv, mes_aniv, ano_atual)
    dias_hoje = dias_desde_inicio_ano(dia_atual, mes_atual, ano_atual)

    if dias_aniversario == dias_hoje:
        return "Parabéns! Hoje é seu aniversário!"
    elif dias_aniversario > dias_hoje:
        return f"Faltam {dias_aniversario - dias_hoje} dias para o seu aniversário."
    else:
        return f"Já se passaram {dias_hoje - dias_aniversario} dias desde o seu aniversário."

dia_aniv = int(input("Digite o dia do seu aniversário: "))
mes_aniv = int(input("Digite o mês do seu aniversário: "))
dia_atual = int(input("Digite o dia de hoje: "))
mes_atual = int(input("Digite o mês atual: "))

resultado = calcular_diferenca(dia_aniv, mes_aniv, dia_atual, mes_atual)
print(resultado)


```

```python
valor1 = float(input("Digite o primeiro valor positivo: "))
valor2 = float(input("Digite o segundo valor positivo: "))
valor3 = float(input("Digite o terceiro valor positivo: "))
valor4 = float(input("Digite o quarto valor positivo:"))
media_ari = (valor1 + valor2 + valor3 + valor4) / 4
media_geo = (valor1 * valor2 * valor3 * valor4) /4
media_har = 4 / (1/valor1 + 1/valor2 + 1/valor3 + 1/valor4)

print(f"\nA média aritmética dos valores é: {media_ari:.2f}")
print(f"\nA média geométrica dos valores é: {media_geo:.2f}")
print(f"\nA média harmônica dos valores é: {media_har:.2f}")
```

```python

n1 = float(input("Digite o 1º número: "))
n2 = float(input("Digite o 2º número: "))
n3 = float(input("Digite o 3º número: "))
n4 = float(input("Digite o 4º número: "))

# Cálculo da média
media = (n1 + n2 + n3 + n4) / 4

# Cálculo da variância
variancia = ((n1 - media)**2 + (n2 - media)**2 + (n3 - media)**2 + (n4 - media)**2) / 4

# Cálculo do desvio padrão
desvio_padrao = variancia ** 0.5

# Saída dos resultados
print(f"\nMédia aritmética: {media:.2f}")
print(f"Variância: {variancia:.2f}")
print(f"Desvio padrão: {desvio_padrao:.2f}")
```
